```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import matplotlib.pyplot as plt
```

```python
XorInput = np.array([[0,0],[0,1],[1,0],[1,1]]).astype('float')
XorOutput = np.array([0,1,1,0]).astype('float')
```

```python
XorInput.shape
```

```python
model = Sequential()
model.add(Dense(2, activation='sigmoid',input_dim = XorInput.shape[1]))
model.add(Dense(1, activation='sigmoid'))
opt = tf.keras.optimizers.RMSprop(learning_rate=0.01)
```

```python
model.compile(optimizer=opt , loss = "binary_crossentropy", metrics = ['mse', 'accuracy'])
```

```python
history = model.fit(XorInput, XorOutput, epochs = 5000, verbose = 1, batch_size = 4)
```

```python
model.predict(XorInput)
```

```python
plt.plot(history.history['mse'])
```
