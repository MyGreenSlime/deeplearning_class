```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import matplotlib.pyplot as plt
```

```python
XorInput = np.array([[0,0],[0,1],[1,0],[1,1]],dtype='float32')
XorOutput = np.array([0,1,1,0], dtype='float32')
```

```python
XorInput.shape
```

```python
import tensorflow.keras.backend as kb
def custom_loss(y_actual,y_pred):
    custom_loss=kb.square(y_actual-y_pred)
    return custom_loss
```

```python
model = Sequential()
model.add(Dense(2, activation='sigmoid',input_dim = XorInput.shape[1]))
model.add(Dense(1, activation='sigmoid'))
opt = tf.keras.optimizers.RMSprop(learning_rate=0.01)
```

```python
model.compile(optimizer=opt , loss = custom_loss, metrics = ['mse', 'accuracy'])
```

```python
history = model.fit(XorInput, XorOutput, epochs = 10000, verbose = 1, batch_size = 1)
```

```python
model.predict(XorInput)
```

```python
plt.plot(history.history['mse'])
```
